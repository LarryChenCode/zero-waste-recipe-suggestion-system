## Import libraries

```python
import pandas as pd
```

## Load data

```python
# Load data
cuisine_df = pd.read_csv('../data/RAW_recipes_with_one_cuisine_replaced_ingr.csv')
cuisine_df.head()
```

## Data preprocessing

```python
# Convert Cuisine_Tags to a single string per row
cuisine_df['Cuisine_Tags_str'] = cuisine_df['Cuisine_Tags'].apply(lambda x: ' '.join(eval(x)))

# Convert replaced_ingredients to a single string per row
cuisine_df['replaced_ingredients_str'] = cuisine_df['replaced_ingredients'].apply(lambda x: ' '.join(eval(x)))

# Convert steps to a single string per row
cuisine_df['steps_str'] = cuisine_df['steps'].apply(lambda x: ' '.join(eval(x)))

# Convert tags to a single string per row
cuisine_df['tags_str'] = cuisine_df['tags'].apply(lambda x: ' '.join(eval(x)))
```

```python
cuisine_df.head()
```

```python
# Move the id, Cuisine_Tags_str, replaced_ingredients_str, steps to the front of the dataframe and then the rest of the columns
cuisine_df = cuisine_df[['id', 'name', 'Cuisine_Tags_str', 'replaced_ingredients_str', 'steps_str', 'tags_str'] + [col for col in cuisine_df.columns if col not in ['id', 'name', 'Cuisine_Tags_str', 'replaced_ingredients_str', 'steps_str', 'tags_str']]]
cuisine_df.head()
```

## Save to csv

```python
# Save the dataframe to a csv file
cuisine_df.to_csv('../data/RAW_recipes_cleaned.csv', index=False)
```
