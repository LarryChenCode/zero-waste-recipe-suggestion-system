## Load data and replace ingredients

```python
# Load the datasets
df_main_w_one_cuisine = pd.read_csv('../data/RAW_recipes_with_one_cuisine.csv')  # Replace with the actual file path
df_main_wo_one_cuisine = pd.read_csv('../data/RAW_recipes_without_one_cuisine.csv')
df_mapping = pd.read_csv('../data/ingr_map.csv')  # Replace with the actual file path

# Create a dictionary to map raw ingredients to replaced ingredients
mapping_dict = pd.Series(df_mapping.replaced.values, index=df_mapping.raw_ingr).to_dict()

# Function to replace ingredients
def replace_ingredients(ingredient_list):
    return [mapping_dict.get(ingredient, ingredient) for ingredient in ingredient_list]

# Apply the mapping to the 'ingredients' column in the main dataset
df_main_w_one_cuisine['replaced_ingredients'] = df_main_w_one_cuisine['ingredients'].apply(eval)  # Convert string representation of list to actual list
df_main_w_one_cuisine['replaced_ingredients'] = df_main_w_one_cuisine['replaced_ingredients'].apply(replace_ingredients)

# Apply the mapping to the 'ingredients' column in the main dataset
df_main_wo_one_cuisine['replaced_ingredients'] = df_main_wo_one_cuisine['ingredients'].apply(eval)  # Convert string representation of list to actual list
df_main_wo_one_cuisine['replaced_ingredients'] = df_main_wo_one_cuisine['replaced_ingredients'].apply(replace_ingredients)
```

## Data preprocessing

```python
## Load data
cuisine_df_w_o_c = pd.read_csv('../data/RAW_recipes_with_one_cuisine_replaced_ingr.csv')
cuisine_df_w_o_c.head()
```

```python
cuisine_df_wo_o_c = pd.read_csv('../data/RAW_recipes_without_one_cuisine_replaced_ingr.csv')
cuisine_df_wo_o_c.head()
```

```python
# Convert Cuisine_Tags to a single string per row
cuisine_df_w_o_c['Cuisine_Tags_str'] = cuisine_df_w_o_c['Cuisine_Tags'].apply(lambda x: ' '.join(eval(x)))

# Convert replaced_ingredients to a single string per row
cuisine_df_w_o_c['replaced_ingredients_str'] = cuisine_df_w_o_c['replaced_ingredients'].apply(lambda x: ' '.join(eval(x)))

# Convert steps to a single string per row
cuisine_df_w_o_c['steps_str'] = cuisine_df_w_o_c['steps'].apply(lambda x: ' '.join(eval(x)))

# Convert tags to a single string per row
cuisine_df_w_o_c['tags_str'] = cuisine_df_w_o_c['tags'].apply(lambda x: ' '.join(eval(x)))
```

```python
# Move the id, Cuisine_Tags_str, replaced_ingredients_str, steps to the front of the dataframe and then the rest of the columns
cuisine_df_w_o_c = cuisine_df_w_o_c[['id', 'name', 'Cuisine_Tags_str', 'replaced_ingredients_str', 'steps_str', 'tags_str'] + [col for col in cuisine_df_w_o_c.columns if col not in ['id', 'name', 'Cuisine_Tags_str', 'replaced_ingredients_str', 'steps_str', 'tags_str']]]
cuisine_df_w_o_c.head()
```

```python
# Convert Cuisine_Tags to a single string per row
cuisine_df_wo_o_c['Cuisine_Tags_str'] = cuisine_df_wo_o_c['Cuisine_Tags'].apply(lambda x: ' '.join(eval(x)))

# Convert replaced_ingredients to a single string per row
cuisine_df_wo_o_c['replaced_ingredients_str'] = cuisine_df_wo_o_c['replaced_ingredients'].apply(lambda x: ' '.join(eval(x)))

# Convert steps to a single string per row
cuisine_df_wo_o_c['steps_str'] = cuisine_df_wo_o_c['steps'].apply(lambda x: ' '.join(eval(x)))

# Convert tags to a single string per row
cuisine_df_wo_o_c['tags_str'] = cuisine_df_wo_o_c['tags'].apply(lambda x: ' '.join(eval(x)))
```

```python
# Move the id, Cuisine_Tags_str, replaced_ingredients_str, steps to the front of the dataframe and then the rest of the columns
cuisine_df_wo_o_c = cuisine_df_wo_o_c[['id', 'name', 'Cuisine_Tags_str', 'replaced_ingredients_str', 'steps_str', 'tags_str'] + [col for col in cuisine_df_wo_o_c.columns if col not in ['id', 'name', 'Cuisine_Tags_str', 'replaced_ingredients_str', 'steps_str', 'tags_str']]]
cuisine_df_wo_o_c.head()
```

## Save to CSV
